# Analysis of aggregated variants

```python
import pandas as pd
```

Read in the data:

```python
agg_variants_file = 'results/aggregated_variants/all_samples.csv'

agg_variants = (
    pd.read_csv(agg_variants_file)
    .assign(mutation=lambda x: x['REF'] + x['POS'].astype(str) + x['ALT'])
    .rename(columns={'POS': 'site',
                     'TOTAL_DP': 'depth',
                     'ALT_FREQ': 'frequency'})
    [['name', 'source', 'run', 'site', 'mutation', 'depth', 'frequency']]
    )
```

Any mutations at sites in likely progenitor?

```python
sites = [8782, 18060, 28144, 29095]

agg_variants.query('site in @sites')
```

What are the consensus mutations?

```python
pd.set_option('display.max_rows', None)

(agg_variants
 .query('frequency > 0.5')
 .sort_values(['name', 'site'])
 .groupby(['name', 'mutation'], sort=False)
 .aggregate(n_runs_with=pd.NamedAgg('frequency', 'count'))
 )
```
